```python
from pandas import Series, DataFrame
import pandas as pd
from sklearn.decomposition import PCA
from sklearn import preprocessing
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

array_Fi = pd.read_csv('X_loadings_792.csv', delimiter=';', decimal='.', names=[i for i in range(10)]) #Открываем данные весов
array_Z = pd.read_csv('X_reduced_792.csv', delimiter=';', decimal='.', names=[i for i in range(10)]) #Открываем данные счетов

array_Fi = np.array(array_Fi) #Переводим данные в матричный вид 
array_Z = np.array(array_Z) #Переводим данные в матричный вид 
array_Fi_tr = array_Fi.transpose() #Находим фи транспонированное

E = np.dot(array_Fi, array_Fi_tr) #Находим E

F = np.dot(array_Z, array_Fi_tr) #Находим исходные данные F

plt.imshow(F) #Визуализируем данные
```
